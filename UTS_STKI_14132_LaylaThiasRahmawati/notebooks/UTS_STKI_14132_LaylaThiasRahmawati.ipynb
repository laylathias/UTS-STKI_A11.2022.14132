## Soal 01 ‚Äì Dasar Sistem Temu Kembali Informasi (IR)
### 1. Definisi STKI dan Perbedaannya dengan Database Retrieval
Sistem Temu Kembali Informasi (STKI) atau Information Retrieval (IR) adalah bidang yang mempelajari bagaimana komputer dapat menemukan dokumen atau informasi yang paling relevan berdasarkan kata kunci yang diberikan pengguna. STKI tidak hanya bertujuan menemukan dokumen yang mengandung kata tertentu, tetapi juga menentukan tingkat relevansi dokumen terhadap kebutuhan pengguna.

Perbedaan utama antara STKI dan Database Retrieval adalah sebagai berikut:
a. Sifat Data
Database Retrieval: bekerja dengan data terstruktur (tabel, kolom, angka).
STKI: bekerja dengan data tidak terstruktur seperti teks panjang, artikel, berita, review, dan web page.

b. Metode Pencarian
- Database Retrieval: pencarian berdasarkan logika pasti (exact match) melalui query SQL.
Misalnya: SELECT * FROM tabel WHERE nama = "Budi".
- STKI: pencarian berbasis makna dan relevansi.
Misalnya pencarian ‚Äúmakan enak‚Äù tidak harus mencari dokumen yang mengandung kedua kata secara persis, tetapi yang paling mendekati makna tersebut.

c. Peran Index & Ranking
Dalam STKI, index berfungsi seperti daftar isi cerdas yang memungkinkan sistem menemukan dokumen dengan cepat. Dua jenis index utama:
Inverted Index
Menyimpan hubungan:
kata ‚Üí daftar dokumen yang mengandung kata tersebut.
Digunakan pada Boolean Retrieval.
Index TF‚ÄìIDF
Menyimpan bobot kemunculan kata dalam dokumen.
Digunakan pada Vector Space Model (VSM).

Karena banyak dokumen bisa cocok dengan satu query, sistem perlu memberi ranking dokumen berdasarkan tingkat relevansi. Ranking dihitung menggunakan cosine similarity pada VSM.

### 2. Garis Besar Arsitektur Search Engine Klasik
Arsitektur search engine klasik seperti Google search engine memiliki beberapa komponen kunci.
Alur utama search engine:

**1. Input Query**
Pengguna mengetik kata kunci, misalnya: ‚Äúmakanan enak pemalang‚Äù.

**2. Preprocessing Query**
Query dibersihkan seperti dokumen: lowercase, tokenizing, stopword removal, stemming

**3. Retrieval Stage**
Sistem mencocokkan query dengan dokumen menggunakan:
Boolean Retrieval (cocok / tidak)
Vector Space Model (menghitung skor kemiripan)

**4. Ranking**
Pada VSM, dokumen diurutkan dari skor tertinggi ke terendah.

**5. Presenting Results**
Sistem menampilkan daftar dokumen, judul, snippet (cuplikan 120 karakter), dan skor relevansi.

### 3. Sketsa Arsitektur IR Klasik (Boolean & VSM)
**A. Arsitektur Boolean Retrieval**
Dokumen Mentah ‚Üí Preprocessing ‚Üí Inverted Index
                               ‚Üë
                               ‚îÇ
                          Query Boolean
                               ‚îÇ
                               ‚Üì
                           Boolean Match

- Boolean mencari dokumen berdasarkan operator AND, OR, NOT.
- Hasilnya berupa set of documents tanpa ranking.
- Cocok untuk pencarian biner: cocok/tidak cocok.

**B. Arsitektur Vector Space Model (VSM)**
Dokumen ‚Üí Preprocessing ‚Üí TF ‚Üí DF ‚Üí IDF ‚Üí TF-IDF ‚Üí Vektor Dokumen
                                                     ‚Üë
                                                     ‚îÇ
                                   Preprocess Query ‚Üí TF-IDF Query
                                                     ‚îÇ
                                                     ‚Üì
                                            Perhitungan Cosine Similarity
                                                     ‚Üì
                                             Ranking Dokumen

- Setiap dokumen diubah menjadi vektor angka (TF-IDF).
- Query juga diubah menjadi vektor.
- Kemiripan dihitung menggunakan cosine similarity.
- Hasil berupa daftar dokumen yang diurutkan relevansinya.
